```python
import os

class BankSystem:
    def __init__(self, filename="accounts.txt"):
        # File path diya h
        self.filename = os.path.join(os.path.expanduser("~"), "Desktop", filename)

        # nayi file create krne ke liye 
        if not os.path.exists(self.filename):
            with open(self.filename, "w") as f:
                f.write("")

    def createAccount(self):
        print("Create a New Account:")
        account_no = input("Enter the Account Number: ")
        name = input("Enter the name of user: ")
        acc_type = input("Enter the account type (c for current / s for savings): ").lower()
        deposit = int(input("Enter the initial amount (>=500 for new account): "))
        password = input("Set your account password: ")

        if deposit >= 500:
            with open(self.filename, "a") as f:
                f.write(f"{account_no},{name},{acc_type},{deposit},{password}\n")
            print("\nAccount created successfully!")
        else:
            print("\nInitial deposit must be at least 500!")

    def login(self):
        print("Login to Your Account:")
        account_no = input("Enter your Account Number: ")
        password = input("Enter your password: ")

        try:
            with open(self.filename, "r") as f:
                accounts = f.readlines()
        except FileNotFoundError:
            print("Account file is missing. Please create an account first.")
            return None

        for account in accounts:
            acc_no, name, acc_type, deposit, acc_password = account.strip().split(",")
            if acc_no == account_no and acc_password == password:
                print(f"\nWelcome, {name}!")
                return acc_no

        print("\nInvalid account number or password!")
        return None

    def displayAccount(self, account_no):
        print("\nAccount Details:")
        try:
            with open(self.filename, "r") as f:
                accounts = f.readlines()
        except FileNotFoundError:
            print("Account file is missing. Please create an account first.")
            return

        for account in accounts:
            acc_no, name, acc_type, deposit, _ = account.strip().split(",")
            if acc_no == account_no:
                print(f"Account Number: {acc_no}")
                print(f"Name: {name}")
                print(f"Account Type: {'Current' if acc_type == 'c' else 'Savings'}")
                print(f"Deposit: {deposit}")
                return

        print("Account not found!")

    def viewAllAccounts(self):
        print("\n--- All Accounts ---")
        try:
            with open(self.filename, "r") as f:
                accounts = f.readlines()
                if not accounts:
                    print("No accounts found.")
                    return

                for account in accounts:
                    acc_no, name, acc_type, deposit, _ = account.strip().split(",")
                    print(f"Account Number: {acc_no}, Name: {name}, Type: {'Current' if acc_type == 'c' else 'Savings'}, Deposit: {deposit}")
        except FileNotFoundError:
            print("Account file is missing. Please create an account first.")
if __name__ == "__main__":
    system = BankSystem()

    while True:
        print("\n--- Bank System Menu ---")
        print("1. Create Account")
        print("2. Login")
        print("3. View All Accounts")
        print("4. Exit")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            system.createAccount()
        elif choice == 2:
            logged_in_account = system.login()
            if logged_in_account:
                system.displayAccount(logged_in_account)
        elif choice == 3:
            system.viewAllAccounts()
        elif choice == 4:
            print("Exiting the system...")
            break
        else:
            print("Invalid choice. Please try again.")
```
